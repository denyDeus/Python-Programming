```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_performance.csv")
```

```python
data["Average"] = (
    data["math"] +
    data["science"] +
    data["english"] +
    data["social_studies"]
) / 4
```

```python
mean_score = data["Average"].mean()

median_score = data["Average"].median()

highest_score = data["Average"].max()

lowest_score = data["Average"].min()
```

```python
print("Student Performance Analysis")
print("-" * 30)

print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score:.2f}")
print(f"Highest Score: {highest_score:.2f}")
print(f"Lowest Score: {lowest_score:.2f}")
```

```python
top_student = data.loc[data["Average"].idxmax()]
lowest_student = data.loc[data["Average"].idxmin()]

print(f"\nTop Student: {top_student['name']} ({top_student['Average']:.2f})")
print(f"Lowest Student: {lowest_student['name']} ({lowest_student['Average']:.2f})")
```

```python
plt.figure(figsize=(10,5))
plt.bar(data["name"], data["Average"])

plt.title("Average Score by Student")
plt.xlabel("Students")
plt.ylabel("Average Score")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    data["Average"],
    labels=data["name"],
    autopct="%1.1f%%"
)

plt.title("Student Performance Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(data["Average"], bins=5)

plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Frequency")

plt.show()
```
